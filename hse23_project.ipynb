```python
# Ищем лучшие e-value
import pandas as pd
import numpy as np

histones = ["H2A", "H2B", "H3", "H4"]
proteomes = ["human", "mouse", "zebrafish", 
             "drosophila", "c_elegans", 
             "ciliate", "yeast", 
             "methanocaldococcus", "thermococcus", 
             "e_coli", "tuberculosis"]

df = pd.DataFrame(columns=["name"] + proteomes)
df
```

```python
def add_crow_to_df(df, proteomes, prefix="", row_name = "name", format="blast"):
  if prefix != "":
    prefix = prefix + "_"
  evalues = []
  for p in proteomes:
    file = open(f"{prefix}{p}.{format}")
    lines = file.readlines()
    if format == "blast":
      evalue = float(lines[5].split("\t")[-2]) if len(lines) > 5 else 1
    else:
      evalue = float(lines[0].split("\t")[-2]) if len(lines) > 0 else 1
    evalues.append(evalue)
    file.close()
  df.loc[df.shape[0]] = [row_name] + evalues
  return df
```

```python
for h in histones:
  df = add_crow_to_df(df, proteomes, prefix=h, row_name=h, format="txt")
```

```python
df = add_crow_to_df(df, proteomes, prefix="", row_name="AICDA", format="blast")
```

```python
# Получили таблицу с e-value, теперь считаем таблицу с -log(e-value)

df_log = -np.log(df)
df_log
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.title("Тепловая карта")
heatmap = sns.heatmap(df_log)
heatmap.set(xlabel="proteomes", ylabel="histones");
```
